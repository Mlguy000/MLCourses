```python
# normalize value ranges
photo = photo / 255
rows = photo.shape[0]
cols = photo.shape[1]
# reshape the array
X = photo.reshape(rows*cols, 3)
X.shape
```

```python
NbColours = 32
```

```python
from sklearn.cluster import KMeans
k_colors = KMeans(n_clusters=NbColours)
k_colors.fit(np.array(X))
```

```python
photo0 = k_colors.cluster_centers_[k_colors.labels_]
photo0 = photo0.reshape(photo.shape)
photo0 = np.asarray(photo0*255, dtype=np.uint8)
```

```python
plt.imshow(photo0)
```

```python
io.imsave("compressed.png", photo0)
```
